````python
import pandas as pd
import numpy as np
````

````python
#reading the csv file into pandas
dirty = pd.read_csv("recweekdata.csv")
dirty
````

````python
# cleaning the month column
month_dict = {"1":"January","2":"February","3":"March","4":"April","5":"May","6":"June","7":"July","8":"August","9":"September","10":"October","11":"November","12":"December","Jan":"January","Feb":"February","Mar":"March","Apr":"April","May":"May","Jun":"June","Jul":"July","Aug":"August","Sep":"September","Oct":"October","Nov":"November","Dec":"December"}

dirty.loc[:,'birthmonth'] = dirty.loc[:,'birthmonth'].str[0:3].apply(lambda x: month_dict[x])
#this will take the string integer as input and replace it with the corresponding string month
````

````python
# cleaning the year column
year_dict = {"97":"1997","98":"1998","99":"1999","1997":"1997","1998":"1998","1999":"1999","2000":"2000","2001":"2001"}

dirty["birthyear"] = dirty.birthyear.astype(str)
dirty.loc[:,'birthyear'] = dirty.loc[:,'birthyear'].str[0:4].apply(lambda x: year_dict[x])
````

````python
# cleaning the courses column
course_dict = {"ME":"BS Management Engineering","ITE":"BS Information Technology Entrepreneurship","CTM":"BS Communications Technology","MGT":"BS Management","LM":"BS Legal Managemnet","BS ME":"BS Management Engineering","BS ITE":"BS Information Technology Entrepreneurship","BS CTM":"BS Communications Technology","BS MGT":"BS Management","BS LM":"BS Legal Management","Management Engineering":"BS Management Engineering","Information Technology Entrepreneurship":"BS Information Technology Entrepreneurship","Communications Technology":"BS Communications Technology","Management":"BS Management","Legal Management":"BS Legal Management"}

dirty.loc[:,'course'] = dirty.loc[:,'course'].str[0:40].apply(lambda x: course_dict[x])
````

````python
# cleaning the phone number column
import re
number_dict = {"917":"Globe","927":"Globe","918":"Smart"}

dirty.loc[:,"phonenumber"] = dirty.loc[:,"phonenumber"].apply(lambda x: re.sub(r"[+]63[ ]","0", x) ) 
#this is to remove the +63 so that the str[0:3] can be consistent
dirty.loc[:,"phonenumber"] = dirty.loc[:,"phonenumber"].apply(lambda x: re.sub(r"^0","", x) )
#dirty

dirty.loc[:,'phonenumber'] = dirty.loc[:,'phonenumber'].str[0:3].apply(lambda x: number_dict[x])
````

# Required Summary Statistics

#### These were the questions from 3A:

1. How many people were born in each month?
2. How many people were born on each day?
3. How many people were born in each year?
4. How many people come from each course?
5. How many people have numbers from Globe (0917 or 0927) and Smart (0918)? *Note: it's up to you how to format this output*

#### Additional Questions:
* How many Globe subscribers are: Male? Female?
* How many Smart subscribers are: Male? Female?

Answer these questions in an html file called **stats3b.html**. The format of the html is as follows (in template form, so you can use this as format string in your program, though feel free to modify with variable names if you find it easier to do so):


```
<html>
    <body>
         <h2>Summary Statistics</h2>
         <table>
             <tr><th>Stat</th><th>Value</th></tr>
             <tr><td>People born in each month</td><td>{}</td></tr>
             <tr><td>People born on each day</td><td>{}</td></tr>
             <tr><td>People born in each year</td><td>{}</td></tr>
             <tr><td>People coming from each course</td><td>{}</td></tr>
             <tr><td>People having numbers from Globe and Smart</td><td>{}</td></tr>
             <tr><td>Male Globe subscribers:</td><td>{}</td></tr>
             <tr><td>Female Globe subscribers:</td><td>{}</td></tr>
             <tr><td>Male Smart subscribers:</td><td>{}</td></tr>
             <tr><td>Female Smart subscribers:</td><td>{}</td></tr>
         </table>
    </body>
</html>
```

````python
stats_html = """
<html>
    <body>
         <h1>Summary Statistics</h1>
         
             <h3>People born in each month</h3>
                 <table>
                     <tr><th>Month</th><th>Count</th></tr>
                     {month}
                 </table>
         
             <h3>People born on each day</h3>
                 <table>
                     <tr><th>Day</th><th>Count</th></tr>
                     {day}
                 </table>
         
             <h3>People born on each year</h3>
                 <table>
                     <tr><th>Year</th><th>Count</th></tr>
                     {year}
                 </table>
         
             <h3>People coming from each course</h3>
                 <table>
                     <tr><th>Course</th><th>Count</th></tr>
                     {course}
                 </table>
         
            <h3>People having numbers from Globe and Smart</h3>
                 <table>
                     <tr><th>Carrier</th><th>Count</th></tr>
                     {carrier}
                 </table>
                 
            <h3>How many Globe subscribers are: Male? Female?</h3>
                 <table>
                     <tr><th>Carrier</th><th>Gender</th><th>Count</th></tr>
                     {gender_carrier}
                 </table>
    </body>
</html>
"""

month_tally_table = ""
month_tally = month_clean.loc[:,["month","lastname"]].apply(lambda x: "<tr><td>{}</td><td>{}</td></tr>".format(x[0],str(x[1])), axis=1)
for month in month_tally:
    month_tally_table += month

day_tally_table = ""
day_tally = day_clean.loc[:,["day","lastname"]].apply(lambda x: "<tr><td>{}</td><td>{}</td></tr>".format(x[0],str(x[1])), axis=1)
for day in day_tally:
    day_tally_table += day
    
year_tally_table = ""
year_tally = year_clean.loc[:,["year","lastname"]].apply(lambda x: "<tr><td>{}</td><td>{}</td></tr>".format(x[0],str(x[1])), axis=1)
for year in year_tally:
    year_tally_table += year
    
course_tally_table = ""
course_tally = course_clean.loc[:,["course","lastname"]].apply(lambda x: "<tr><td>{}</td><td>{}</td></tr>".format(x[0],str(x[1])), axis=1)
for course in course_tally:
    course_tally_table += course

carrier_tally_table = ""
carrier_tally = phone_clean.loc[:,["carrier","lastname"]].apply(lambda x: "<tr><td>{}</td><td>{}</td></tr>".format(x[0],str(x[1])), axis=1)
for carrier in carrier_tally:
    carrier_tally_table += carrier

gender_tally_table = ""
gender_tally = merged_phone_gender.loc[:,["phonenumber","gender","lastname"]].apply(lambda x: "<tr><td>{}</td><td>{}</td><td>{}</td></tr>".format(x[0],x[1],str(x[2])), axis=1)
for gender in gender_tally:
    gender_tally_table += gender
    
with open('stats3b.html',"w+") as file:
    content_html = stats_html.format(month = month_tally_table, day = day_tally_table, year = year_tally_table, course = course_tally_table, carrier = carrier_tally_table, gender_carrier = gender_tally_table)
    file.write(content_html)
    
````
